## Using product famille as a target
We can use the 4 main product familles for which we have reasonable number of products.

```python
familles = ['Lingerie', 'Homewear', 'Chaussettes, Collants et Bas', 'Spécial bébé']
```

```python
df2 = df[df['nomenclature_famille'].isin(familles)].copy()
len(df2)
```

```python
df2['nomenclature_famille'].value_counts()
```

```python
df2[base_cols].sample(10)
```

```python
df3 = df2.reset_index()
df3.head()
```

```python
#Map famille to a family_code
def fam_map(x):
    return familles.index(x)

fam_code = fam_map('Homewear')
```

```python
#Basic test
test_fam = familles[1]
fam_code = fam_map(test_fam)
print(f'{test_fam} => {fam_code}')
```

```python
fam = 'nomenclature_famille'
df3["famille_code"] = df3[fam].map(fam_map)
df3.head()
```

```python
# Store result file
out_file_name = './data_test/dim_product_clean.csv'
df3.to_csv(out_file_name, index=False)
```
